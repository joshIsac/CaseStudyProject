```python
#extracting image from the URL 
import os 
import urllib.request
from PIL import Image
from tqdm import tqdm
image_dir=r'D:\\dataset for lab programs\\Case Study Project\\images_1'
os.makedirs(image_dir, exist_ok=True)
```

```python
from urllib.parse import urlparse
# Function to download and preprocess images
def download_and_preprocess_images(data, limit=1000):
    image_paths = []
    for idx, url in tqdm(enumerate(data['Image_URL']), total=limit):
        try:
            # Parse URL and sanitize it
            parsed_url = urlparse(url)
            sanitized_url = parsed_url.geturl()
            
            # Define image path
            image_path = os.path.join(image_dir, f'image_{idx}.jpg')
            
            # Check if image already exists
            if os.path.exists(image_path):
                image_paths.append(image_path)
                continue
            
            # Download image
            urllib.request.urlretrieve(sanitized_url, image_path)
            
            # Open and preprocess image
            image = Image.open(image_path)
            image = image.resize((224, 224))  # Resize to a fixed dimension (e.g., 224x224)
            image = np.array(image) / 255.0   # Normalize pixel values
            image_paths.append(image_path)
        except Exception as e:
            print(f"Error downloading image at index {idx}: {e}")
        
        if len(image_paths) >= limit:
            break  # Break out of the loop after downloading the specified number of images
    
    return image_paths

# Download and preprocess images (limiting to 5 images)
image_paths = download_and_preprocess_images(data, limit=1000)
```

```python
# Function to display images (up to 10 images)
def display_images(image_paths):
    num_images = min(len(image_paths), 10)  # Limit to 10 images if more than 10 are available
    plt.figure(figsize=(15, 6))
    for i in range(num_images):
        plt.subplot(2, 5, i + 1)  # Arrange images in a 2x5 grid
        image = Image.open(image_paths[i])
        plt.imshow(image)
        plt.axis('off') 
    plt.show()

# Display the downloaded images
display_images(image_paths)
```

```python
import numpy as np
import pandas as pd
from transformers import BertTokenizer, TFBertModel
from keras.applications import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
```

```python
# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = TFBertModel.from_pretrained('bert-base-uncased')
```

```python
# Function to extract textual features using BERT
def extract_textual_features(text_data):
    input_ids = tokenizer(text_data.tolist(), return_tensors='tf', padding=True, truncation=True)['input_ids']
    textual_features = bert_model(input_ids)[0][:, 0, :].numpy()  # Extract features from the [CLS] token
    return textual_features
```

```python
# Limiting the number of textual features to 1000
text_data = data['Brand'][:1000] + ' ' + data['Description'][:1000]

textual_features = extract_textual_features(text_data)
print("Textual features shape:", textual_features.shape)
```

```python
#extracting visual features 
# Load pre-trained VGG16 model
vgg_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

from keras.preprocessing import image

# Function to extract visual features using VGG16
def extract_visual_features(image_paths, limit=1000):
    visual_features = []
    count = 0
    for path in image_paths:
        if count >= limit:
            break
        img = image.load_img(path, target_size=(224, 224))
        img = image.img_to_array(img)
        img = preprocess_input(img)
        img = np.expand_dims(img, axis=0)
        features = vgg_model.predict(img)
        visual_features.append(features.flatten())
        count += 1
    visual_features = np.array(visual_features)
    return visual_features
```
